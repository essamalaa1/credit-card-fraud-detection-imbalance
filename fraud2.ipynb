```python
# Split the data into training and test sets with stratification
from sklearn.model_selection import train_test_split
train_df_unbalanced, test_df_unbalanced = train_test_split(df, test_size=0.3,
                                                             stratify=df['Class'], random_state=42)

print("Train class distribution:\n", train_df_unbalanced['Class'].value_counts())
print("\nTest class distribution:\n", test_df_unbalanced['Class'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

X_train = train_df_unbalanced.drop(columns=['Class'])
y_train = train_df_unbalanced['Class']

X_test = test_df_unbalanced.drop(columns=['Class'])
y_test = test_df_unbalanced['Class']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Update the DataFrames with scaled features
train_df_unbalanced = pd.DataFrame(X_train_scaled, columns=train_df_unbalanced.columns.drop('Class'))
train_df_unbalanced['Class'] = y_train.values

test_df_unbalanced = pd.DataFrame(X_test_scaled, columns=test_df_unbalanced.columns.drop('Class'))
test_df_unbalanced['Class'] = y_test.values
```

```python
train_df_unbalanced.describe()
```

```python
test_df_unbalanced.describe()
```

```python
X_train = train_df_unbalanced.drop(columns=['Class'])
y_train = train_df_unbalanced['Class']
```

```python
# 1. SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

train_df_smote = pd.DataFrame(X_train_smote, columns=X_train.columns)
train_df_smote['Class'] = y_train_smote
print("After SMOTE, training class distribution:\n", train_df_smote['Class'].value_counts())
```

```python
# 2. Random Undersampling
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

train_df_rus = pd.DataFrame(X_train_rus, columns=X_train.columns)
train_df_rus['Class'] = y_train_rus
print("After Random Undersampling, training class distribution:\n", train_df_rus['Class'].value_counts())
```

```python
from imblearn.over_sampling import RandomOverSampler
# 3. Random Oversampling
ros = RandomOverSampler(random_state=42)
X_train_ros, y_train_ros = ros.fit_resample(X_train, y_train)
train_df_ros = pd.DataFrame(X_train_ros, columns=X_train.columns)
train_df_ros['Class'] = y_train_ros
print("After Random Oversampling, training class distribution:\n",train_df_ros['Class'].value_counts())
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
X_test_eval = test_df_unbalanced.drop(columns=['Class'])
y_test_eval = test_df_unbalanced['Class']
```

```python
from sklearn.neural_network import MLPClassifier

def evaluate_nn(X_train_data, y_train_data, label):
    print(f"Training samples used: {len(X_train_data)}")

    # Train a simple MLP (2 hidden layers: 64 and 32 units)
    nn_model = MLPClassifier(
        hidden_layer_sizes=(64, 32),
        activation='relu',
        solver='adam',
        max_iter=100,
        random_state=42,
        verbose=False
    )
    nn_model.fit(X_train_data, y_train_data)

    # Predict on the test set
    y_test_pred = nn_model.predict(X_test_eval)
    test_acc = accuracy_score(y_test_eval, y_test_pred)

    print(f"\n====== {label} (NN) ======")
    print(f"Test Accuracy: {test_acc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test_eval, y_test_pred)
    print("Confusion Matrix (Test):")
    print(cm)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="viridis", cbar=True)
    plt.title(f"{label} (NN) - Confusion Matrix")
    plt.xlabel("Predicted Class")
    plt.ylabel("True Class")
    plt.show()

    # Classification Report
    cr_dict = classification_report(y_test_eval, y_test_pred, output_dict=True)
    cr_df = pd.DataFrame(cr_dict).transpose()
    print("Classification Report (Test):")
    print(cr_df)
```

```python
evaluate_nn(
    train_df_unbalanced.drop(columns=['Class']),
    train_df_unbalanced['Class'],
    label="Original Unbalanced"
)
```

```python
evaluate_nn(
    train_df_smote.drop(columns=['Class']),
    train_df_smote['Class'],
    label="SMOTE"
)
```

```python
evaluate_nn(
    train_df_rus.drop(columns=['Class']),
    train_df_rus['Class'],
    label="Under Sampling"
)
```

```python
evaluate_nn(
    train_df_ros.drop(columns=['Class']),
    train_df_ros['Class'],
    label="OverSample"
)
```
